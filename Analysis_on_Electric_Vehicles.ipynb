```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Hello")

df = pd.read_csv("dataset (1).csv")

df.info()
df.head()

sns.set(style="whitegrid")

categorical_cols = ['Make', 'Model', 'State', 'Electric Vehicle Type', 
                    'Clean Alternative Fuel Vehicle (CAFV) Eligibility']

fig, axes = plt.subplots(len(categorical_cols), 1, figsize=(10, 20))

for i, col in enumerate(categorical_cols):
    sns.countplot(data=df, x=col, ax=axes[i], order=df[col].value_counts().index)
    axes[i].set_title(f'Distribution of {col}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

numerical_cols = ['Model Year', 'Electric Range', 'Base MSRP']

df[numerical_cols].hist(bins=20, figsize=(15, 8), layout=(1, 3))
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Make', y='Electric Range', data=df)
plt.title('Electric Range by Make')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Electric Vehicle Type', y='Model Year', data=df)
plt.title('Model Year by Electric Vehicle Type')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='State', hue='Electric Vehicle Type', data=df, order=df['State'].value_counts().index)
plt.title('State-wise Distribution of Electric Vehicle Types')
plt.xticks(rotation=45)
plt.show()
```

```

import pandas as pd
import plotly.express as px

print("Hello ")

df = pd.read_csv("dataset (1).csv")

state_ev_count = df.groupby('State').size().reset_index(name='EV Count')

fig = px.choropleth(state_ev_count, 
                    locations='State', 
                    locationmode='USA-states', 
                    color='EV Count', 
                    hover_name='State', 
                    color_continuous_scale='Viridis', 
                    scope='usa',  
                    labels={'EV Count':'Number of EVs'})

fig.update_layout(title_text='Number of Electric Vehicles by State in the USA',
                  geo_scope='usa')

fig.show()
```

```

df = pd.read_csv("dataset (1).csv")

df.columns

df_grouped = df.groupby(['Model Year' , 'Make']).size().reset_index(name='Count')

if 'Count' not in df_grouped.columns:
    print("Count column not found! check the previous steps.")
else:
    fig = px.bar(df_grouped,
                 x='Count',
                 y='Make',
                 color='Count',
                 animation_frame='Model Year',
                 range_x=[0,df_grouped['Count'].max() * 1.1],
                 title='Year-wise Sales of Each EV Maker',
                 labels={'Count': 'Number of Sales','Make' : 'EV Maker'},
                 orientation='h')

fig.show()
```
